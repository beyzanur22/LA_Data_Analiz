```python
import pandas as pd
import os
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# --- 1. DOSYA YOLLARINI TANIMLA ---
BASE_DIR = r"C:\Users\goktu\workspace\datacops"
OUTPUT_DIR = os.path.join(BASE_DIR, "model_goktug")

# Ger√ßek Cevaplar (Test Seti)
TEST_PATH = os.path.join(BASE_DIR, "test_scaled.csv")

# Senin Olu≈üturduƒüun Tahmin Dosyalarƒ±
SUB_1_PATH = os.path.join(OUTPUT_DIR, "submission_goktug_svm.csv")
SUB_2_PATH = os.path.join(OUTPUT_DIR, "submission_goktug_model2.csv")

# --- 2. DOSYALARI OKU ---
print("Dosyalar okunuyor...")
df_test = pd.read_csv(TEST_PATH)
df_sub1 = pd.read_csv(SUB_1_PATH)
df_sub2 = pd.read_csv(SUB_2_PATH)

# Hedef ve Tahminleri Al (String formatƒ±nda olduƒüundan emin olalƒ±m)
y_true = df_test["Crm Cd"].astype(str)
y_pred_1 = df_sub1["Crm Cd"].astype(str)
y_pred_2 = df_sub2["Crm Cd"].astype(str)

# --- 3. METRƒ∞KLERƒ∞ HESAPLA ---
def get_metrics(model_name, y_true, y_pred):
    return {
        "Model": model_name,
        "Accuracy": accuracy_score(y_true, y_pred),
        "Precision": precision_score(y_true, y_pred, average='weighted', zero_division=0),
        "Recall": recall_score(y_true, y_pred, average='weighted', zero_division=0),
        "F1-Score": f1_score(y_true, y_pred, average='weighted', zero_division=0)
    }

metrics_1 = get_metrics("Linear SVM (ANOVA)", y_true, y_pred_1)
metrics_2 = get_metrics("Ridge Classifier (Tree-Based)", y_true, y_pred_2)

# --- 4. TABLOYU OLU≈ûTUR VE G√ñSTER ---
df_comparison = pd.DataFrame([metrics_1, metrics_2])
df_comparison.set_index("Model", inplace=True)


print("üèÜ MODEL KAR≈ûILA≈ûTIRMA TABLOSU")
display(df_comparison)

excel_path = os.path.join(OUTPUT_DIR, "model_karsilastirma_final.xlsx")
df_comparison.to_excel(excel_path)
print(f"\n‚úÖ Tablo Excel olarak da kaydedildi: {excel_path}")
```
