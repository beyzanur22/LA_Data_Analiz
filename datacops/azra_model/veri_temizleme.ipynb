```python
import pandas as pd

path = "C:\AzraTanriver\Crime_Data_from_2020_to_Present.csv"

df = pd.read_csv(path)

# Boyut ve kolonlar
print(df.shape)
print(df.columns)

# Genel bilgi
df.info()

# İlk satırlar
df.head()
```

```python
missing_counts = df.isnull().sum().sort_values(ascending=False)
missing_ratio = (df.isnull().mean()*100).sort_values(ascending=False)

missing_df = pd.DataFrame({
    "missing_count": missing_counts,
    "missing_ratio_%": missing_ratio.round(2)
})
missing_df.head(20)
```

```python
# 1. Gereksiz sütunları sil
df = df.drop(columns=['Crm Cd 2', 'Crm Cd 3', 'Crm Cd 4', 'Cross Street'])

# 2. Eksik kategorik sütunları doldur
df['Weapon Desc'] = df['Weapon Desc'].fillna('UNKNOWN')
df['Weapon Used Cd'] = df['Weapon Used Cd'].fillna(-1)  # -1 = bilinmeyen
df['Vict Sex'] = df['Vict Sex'].fillna('UNKNOWN')
df['Vict Descent'] = df['Vict Descent'].fillna('UNKNOWN')
df['Premis Desc'] = df['Premis Desc'].fillna('UNKNOWN')

# 3. Kategorik sütunları formatla
df['Vict Sex'] = df['Vict Sex'].str.strip().str.upper()
df['Vict Descent'] = df['Vict Descent'].str.strip().str.upper()
df['Weapon Desc'] = df['Weapon Desc'].str.strip().str.upper()
```

```python
df.isnull().sum().sort_values(ascending=False).head(10)
```

```python
# Tarih dönüşümü (format belirtilmiş)
df['DATE OCC'] = pd.to_datetime(df['DATE OCC'], format='%m/%d/%Y %I:%M:%S %p', errors='coerce')
df['Date Rptd'] = pd.to_datetime(df['Date Rptd'], format='%m/%d/%Y %I:%M:%S %p', errors='coerce')

# Geçersiz (NaT) tarihleri at
df = df.dropna(subset=['DATE OCC', 'Date Rptd'])

# Rapor tarihi olay tarihinden önce olmasın
df = df[df['Date Rptd'] >= df['DATE OCC']]

# 2020 öncesi kayıtları kaldır
df = df[df['DATE OCC'].dt.year >= 2020]
```

```python
df[['DATE OCC', 'Date Rptd']].head()
```

```python
# Rapor tarihi olay tarihinden önceyse sil
df = df[df['Date Rptd'] >= df['DATE OCC']]

# Aşırı ileri tarihli raporları (örneğin 1 yıldan fazla fark varsa) isteğe bağlı kaldırabiliriz
df = df[(df['Date Rptd'] - df['DATE OCC']).dt.days <= 365]

# 2020 öncesi kayıtları kaldır (emin olmak için tekrar)
df = df[df['DATE OCC'].dt.year >= 2020]
#bu aşamada tarih tutarsızlıkları temizliyoruz
```

```python
(df['Date Rptd'] - df['DATE OCC']).describe()
```

```python
import numpy as np

# 0 veya 110'dan büyük yaşları geçersiz say
df['Vict Age'] = df['Vict Age'].mask((df['Vict Age'] <= 0) | (df['Vict Age'] > 110))

# medyan yaşı hesapla
median_age = df['Vict Age'].median()
print("Medyan yaş:", median_age)

# NaN olanları medyanla doldur
df['Vict Age'] = df['Vict Age'].fillna(median_age)
```

```python
before_shape = df.shape

df = df[(df['LAT'] > 33) & (df['LAT'] < 35)]
df = df[(df['LON'] > -119) & (df['LON'] < -117)]

after_shape = df.shape
print("Önce:", before_shape, "Sonra:", after_shape)
```

```python
# DR_NO'ya göre tekrar eden olaylar var mı?
print("Tekrarlı olay sayısı:", df.duplicated(subset=['DR_NO']).sum())

# Varsa sil
df = df.drop_duplicates(subset=['DR_NO'])

# Son durum kontrolü
print(df.shape)
df.info()

# Temizlenmiş versiyonu kaydet
df.to_csv("crime_data_clean_azra.csv", index=False)
```
